```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
```

```python
df =pd.read_csv(r'E:\User Data\Documents\rakamin\EDA\train.csv')
```

```python
df.head()
```

```python
rcParams['figure.figsize']= (10,7)
rcParams['lines.linewidth']= 2.5
rcParams['xtick.labelsize']= 'x-large'
rcParams['ytick.labelsize']= 'x-large'
```

```python
df.info()
```

```python
numericals = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
categoricals = ['Sex', 'Embarked']
```

```python
features = numericals
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.boxplot(y=df[features[i]], color='red', orient='v')
    plt.tight_layout()
```

```python
features = numericals
for i in range(0, len(features)):
    plt.subplot(2, 4, i+1)
    sns.kdeplot(x=df[features[i]], color='red')
    plt.xlabel(features[i])
    plt.tight_layout()
```

```python
features = numericals
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.violinplot(y=df[features[i]], color='red')
    plt.tight_layout()
```

```python
features = categoricals
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.countplot(x=df[features[i]], color='red')
    plt.tight_layout()
```
